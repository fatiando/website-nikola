## Example: Simple 2D tomography

Let's have a look at a simple tomography example using module [fatiando.seismic.srtomo](http://fatiando.readthedocs.org/en/Version0.1/api/seismic.srtomo.html#module-fatiando.seismic.srtomo).
For this example, we assume that seismic waves travel along straight ray paths.
In other words, there is no reflection or refraction of waves when the velocity changes.

Load pylab so that images are embeded in the notebook:

```python
%pylab inline
```

First thing we need is a velocity model.
We'll use the Fatiando a Terra logo image as a basis for this.

```python
import urllib
from fatiando import mesher, utils, seismic
from fatiando.vis import mpl

area = (0, 500000, 0, 500000)
shape = (30, 30)
# Create a mesh where the velocities are defined
model = mesher.SquareMesh(area, shape)
# Fetch the image from the online docs
urllib.urlretrieve('http://fatiando.readthedocs.org/en/latest/_static/logo.png', 'logo.png')
model.img2prop('logo.png', 4000, 10000, 'vp')

mpl.figure(figsize=(7, 5))
mpl.axis('scaled')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity (m/s)')
mpl.m2km()
```

OK, now that we have a velocity model, let's simulate some earthquakes!
We'll generate some random source points (quakes) which will be recorded at a circular array of seismometers (receivers).

```python
quake_locations = utils.random_points(area, 40)
receiver_locations = utils.circular_points(area, 20, random=True)
quakes, seismometers = utils.connect_points(quake_locations, receiver_locations)

mpl.figure(figsize=(7, 5))
mpl.axis('scaled')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity (m/s)')
mpl.points(quake_locations, '*y', label="Sources")
mpl.points(receiver_locations, '^g', label="Receivers")
mpl.m2km()
```

Next, we calculate the time it takes for a ray traveling from the earthquakes to reach the receivers.
Note that the speed at which the ray travels varies depending on which cell it's in.

```python
traveltimes = seismic.ttime2d.straight(model, 'vp', quakes, seismometers)

mpl.figure(figsize=(7, 5))
mpl.axis('scaled')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity (m/s)')
mpl.paths(quakes, seismometers)
mpl.points(quake_locations, '*y', label="Sources")
mpl.points(receiver_locations, '^g', label="Receivers")
mpl.m2km()
```

That gives us a pretty good ray coverage in the middle of the image.
This is a hint that the tomography will be able to recover the middle image, but not the edges.
Let's run it and see!

Note: The tomography actually estimates the slowness (1/velocity) of each cell. This is just so the math is a bit easier.

```python
# Make the mesh for the tomography
mesh = mesher.SquareMesh(area, shape)
# and then run it to estimate the slowness of the cells
slowness, residuals = seismic.srtomo.run(traveltimes, quakes, seismometers, mesh)
# Convert the slowness estimate to velocities and add it the mesh so it can be plotted
mesh.addprop('vp', seismic.srtomo.slowness2vel(slowness))
```

What was that? OK, [don't panic](http://en.wikipedia.org/wiki/Don%27t_Panic_%28The_Hitchhiker%27s_Guide_to_the_Galaxy%29#Don.27t_Panic)!
That just means that we can't estimate the velocities (or slowness) of the cells based on the traveltime data **alone**.
But why is that?
If you have a look at the borders of the image above, you'll notice that some cells have absolutely **zero rays** passing through them.
So can we know the velocity, if we don't know how long it takes something to pass through it?
Well, in short, we can't.

But all is not lost!
Regularization comes to the rescue!
Since the data doesn't tell us what the velocity should be like, we have to use input our own knowledge into the tomography.
That knowlegde is usually refered to as "a priori information".

In this case, let's assume that we know that the velocity varies smoothly.
There is a way to input that information through [Tikhonov regularization](http://en.wikipedia.org/wiki/Tikhonov_regularization).
I won't get into the details of that (but it will be the subject of a future post).
Smoothness is already implemented in [fatiando.seismic.srtomo](http://fatiando.readthedocs.org/en/Version0.1/api/seismic.srtomo.html#module-fatiando.seismic.srtomo)
and can be imposed using the "smooth" keyword argument.
It is a floating point number that controls just how smooth we want the estimated slowness to be.
Let's apply some smoothness and see what happens!

```python
# Make the mesh for the tomography
mesh = mesher.SquareMesh(area, shape)
# and then run it to estimate the slowness of the cells
slowness, residuals = seismic.srtomo.run(traveltimes, quakes, seismometers, mesh, smooth=0.1)
# Convert the slowness estimate to velocities and add it the mesh so it can be plotted
mesh.addprop('vp', seismic.srtomo.slowness2vel(slowness))

mpl.figure(figsize=(14, 5))
mpl.subplot(1, 2, 1)
mpl.title('True')
mpl.axis('scaled')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity (m/s)')
mpl.m2km()
mpl.subplot(1, 2, 2)
mpl.title('Estimated')
mpl.axis('scaled')
mpl.squaremesh(mesh, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity (m/s)')
mpl.m2km()
```

Much better!
Now what is interesting about this result is that the borders are all smudged.
This is because the data supply **no** information there.
All that governs the estimate in those places is the regularization.
And the regularization tells it to be smooth, so it is smooth!
